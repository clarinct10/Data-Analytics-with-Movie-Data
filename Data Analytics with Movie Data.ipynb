## Exploratory Data Analysis

#### Question 1: Number of votes per year for movie?

```python
# Return the index of columns
data.columns
```

```python
data.groupby('Year')['Votes'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Year',y='Votes',data=data)
plt.title("Votes By Year")
plt.show()
```

The highest number of votes was obtained in 2012 followed by 2008 as the second most votes.

#### Question 2: Number of revenue per year for movie?

```python
data.groupby('Year')['Revenue (Millions)'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Year',y='Revenue (Millions)',data=data)
plt.title("Revenue By Year")
plt.show()
```

The highest income was obtained in 2009 amounted to 112.601277

#### Question 3: Shortest and Longest  runtime for movie?

```python
# To get n least values from a data frame
data.nsmallest(10,'Runtime (Minutes)')
```

```python
# To get n largest values from a data frame 
data.nlargest(10,'Runtime (Minutes)')
```

```python
top10_largest =data.nlargest(10,'Runtime (Minutes)')[['Title', 'Runtime (Minutes)']].set_index('Title')
top10_largest
```

```python
sns.barplot(x='Runtime (Minutes)',y=top10_largest.index,data=top10_largest)
plt.show()
```

The shortest film runtime is Ma vie de Courgette with 66 minutes
and The longest film runtime is Grindhouse with 191 minutes.

#### Question 4: Number of movie per year?

```python
# Returns a Series containing counts of unique values
data['Year'].value_counts()
```

```python
sns.countplot(x='Year',data=data)
plt.title('Number of Movies Per Year')
plt.show()
```

The highest number of films was obtained in 2016 which experienced a significant increase from the previous year.

#### Question 5: Which movie has the highest rating?

```python
top10_rating=data.nlargest(10,'Rating')[['Title', 'Rating', 'Director']].set_index('Title')
top10_rating
```

```python
sns.barplot(x='Rating', y=top10_rating.index, data=top10_rating, hue='Director',dodge=False)
plt.legend(bbox_to_anchor=(1.05,1), loc=2)
plt.show()
```

The film with the highest rating is The Dark Knight followed by the film Inception as the second highest rating with the same director named Christopher Nolan.

#### Question 6: Which movie has the highest revenue?

```python
data[data['Revenue (Millions)'].max()==data['Revenue (Millions)']]['Title']
```

```python
top_10=data.nlargest(10,'Revenue (Millions)')[['Title','Revenue (Millions)']].set_index('Title')
top_10
```

```python
sns.barplot(x='Revenue (Millions)', y=top_10.index, data=top_10)
plt.title("Top Highest Revenue Movie Titles")
plt.show()
```

The highest revenue film was Star Wars: Episode VII - The Force Awakens with 936.63 Millions.

#### Question 7: Number of ratings each year? Effect of rating on revenue?

```python
data.groupby('Year')['Rating'].mean().sort_values(ascending=False)
```

```python
sns.scatterplot(x='Rating', y='Revenue (Millions)', data=data)
plt.show()
```

The highest total rating was obtained in 2007. The relationship between ratings and revenue can be seen from the graph shown 
that the higher the film rating, the more it will have an impact on increasing film revenue.
